```python
def func_SHAPIROWILK2(group, 
                        str_method,
                       boo_dropIguales=True
                       ):


    from scipy.stats import shapiro
    #statistic, p_value = wilcoxon(group[str_methodPropuesto], group[str_methodBase], alternative='less')
    statistic, p_value = shapiro(group[str_method])
      
    alpha = 0.05
    if p_value > alpha:
        #print("\x1b[32m\"The data appears to be normally distributed.\"\x1b[0m")
        print("The data appears to be normally distributed. (SHAPIRO)")
    else:
        print("The data significantly deviates from a normal distribution. (SHAPIRO)")
    return(p_value)
```

```python


def func_reemplazarHeader(df_In, 
                          boo_metodoDLoArtProbIn):
    df_out=df_In.rename(columns={
        'kernelShape':r'\textbf{\shortstack{kernel \\ shape }}', 
        
        'delta_Connected_mask_DL_DISCSAVG_meanSE':r'\textbf{\shortstack{Connected \\ delta \\ GT-'+boo_metodoDLoArtProb+r'}}',                                     
        'delta_Connected_mask_CRF_DISCSAVG_meanSE':r'\textbf{\shortstack{Connected \\ delta \\ GT-CRF }}',                                             
        'delta_Connected_mask_CRFROT_DISCSAVG_meanSE':r'\textbf{\shortstack{Connected \\ delta \\ GT-CRFROT }}', 
                
        'delta_Median_mask_DL_DISCSAVG_meanSE':r'\textbf{\shortstack{Median \\ delta \\ GT-'+boo_metodoDLoArtProb+r'}}', 
        'delta_Median_mask_CRF_DISCSAVG_meanSE':r'\textbf{\shortstack{Median \\ delta \\ GT-CRF}}', 
        'delta_Median_mask_CRFROT_DISCSAVG_meanSE':r"\textbf{\shortstack{Median \\ delta \\ GT-CRFROT}}", 
        
        'delta_Sum_mask_DL_DISCSAVG_meanSE':r'\textbf{\shortstack{Sum \\ delta \\ GT-'+boo_metodoDLoArtProb+r'}}', 
        'delta_Sum_mask_CRF_DISCSAVG_meanSE':r'\textbf{\shortstack{Sum \\ delta \\ GT-CRF}}', 
        'delta_Sum_mask_CRFROT_DISCSAVG_meanSE':r'\textbf{\shortstack{Sum \\ delta \\ GT-CRFROT}}', 
        
        "PVAL "+boo_metodoDLoArtProb+">CRF":r'\textbf{\shortstack{PVAL \\ GT-'+boo_metodoDLoArtProb+r' \\ > \\GT-CRF}}',            
        "PVAL "+boo_metodoDLoArtProb+">CRFROT":r'\textbf{\shortstack{PVAL \\ GT-'+boo_metodoDLoArtProb+r' \\ > \\GT-CRFROT}}',            
        "PVAL CRF>CRFROT":r'\textbf{\shortstack{PVAL \\ GT-CRF \\ > \\GT-CRFROT}}',            
        
        }
                       )
    return(df_out)


def func_crearTablaenLatex(df_In):
    int_cantidadColumnas=len(df_In.columns)

    df_aux_kernelshape_str = r"\begin{tabular}"+r"{"+int_cantidadColumnas*"c"+r"}"+ '\n'
    df_aux_kernelshape_str += r"\toprule"+ '\n'
    df_aux_kernelshape_str += ' & '.join(df_In.columns) + ' \\\\\n'  # Include column names
    df_aux_kernelshape_str += r"\midrule" + '\n'
    df_aux_kernelshape_str += df_In.apply(lambda row: ' & '.join(row.astype(str)), axis=1).str.cat(sep=' \\\\\n')+ ' \\\\\n'                    
    df_aux_kernelshape_str += r"\bottomrule" + '\n'
    df_aux_kernelshape_str += r"\end{tabular}"
    return(df_aux_kernelshape_str)

        
def func_negritaparaelminimo_ValoresMINSE(dfIn):
    """
    El df debe ser numerico, no string como en func_negritaparaelminimo_Aplicar
    """
    dfOut=dfIn.copy()
    
    lst_flo_negritaparaelminimo_ValoresMINout=[]
    lst_flo_negritaparaelminimo_ValoresSEout=[]
    lst_str_negritaparaelminimo_Columnas=[
        "delta_Connected_mask_CRF_DISCSAVG_meanSE",
        "delta_Connected_mask_CRFROT_DISCSAVG_meanSE",
        
        #"delta_Median_mask_CRF_DISCSAVG_meanSE",
        #"delta_Median_mask_CRFROT_DISCSAVG_meanSE",        
        
        #"delta_Sum_mask_CRF_DISCSAVG_meanSE",
        #"delta_Sum_mask_CRFROT_DISCSAVG_meanSE"        
                                        ]
    for lst_str_negritaparaelminimo_Columnas_i in lst_str_negritaparaelminimo_Columnas:
        
        min_valueCRF = dfOut[lst_str_negritaparaelminimo_Columnas_i].str.split(' ± ').str[0].astype(float).min()
        min_seCRF = dfOut.loc[dfOut[lst_str_negritaparaelminimo_Columnas_i].str.contains(str(min_valueCRF))][lst_str_negritaparaelminimo_Columnas_i].str.split(' ± ').str[1].values[0]

        
        lst_flo_negritaparaelminimo_ValoresMINout.append(min_valueCRF)
        lst_flo_negritaparaelminimo_ValoresSEout.append(min_seCRF)
    #df_aux_kernelshape_str=df_aux_kernelshape_str.replace(f"{max_valueCRF} ± {max_seCRF}", f"\\textbf{{{max_valueCRF} ± {max_seCRF}}}")
    #df_aux_kernelshape_str=df_aux_kernelshape_str.replace(f"{max_valueCRFROT} ± {max_seCRFROT}", f"\\textbf{{{max_valueCRFROT} ± {max_seCRFROT}}}")
    
    return(lst_flo_negritaparaelminimo_ValoresMINout,
           lst_flo_negritaparaelminimo_ValoresSEout
          )

def func_negritaparaelminimo_Aplicar(
                                    dfIn_str,
                                    lst_flo_negritaparaelminimo_ValoresMINIn,
                                    lst_flo_negritaparaelminimo_ValoresSEIn
                                    ):
    """
    El df debe ser string, no numerico como en func_negritaparaelminimo_Valores
    En este metodo deja dos valores con negrita si hay 2 valores iguales (lo mismo ocurre con mas de 2)
    """
    
    
    for lst_flo_negritaparaelminimo_ValoresMINIn_i, lst_flo_negritaparaelminimo_ValoresSEIn_i in zip(lst_flo_negritaparaelminimo_ValoresMINIn, lst_flo_negritaparaelminimo_ValoresSEIn):
        #print(f"{lst_flo_negritaparaelminimo_ValoresMINIn_i} ± {lst_flo_negritaparaelminimo_ValoresSEIn_i}")
        dfIn_str=dfIn_str.replace(f"{lst_flo_negritaparaelminimo_ValoresMINIn_i} ± {lst_flo_negritaparaelminimo_ValoresSEIn_i}", f"\\textbf{{{lst_flo_negritaparaelminimo_ValoresMINIn_i} ± {lst_flo_negritaparaelminimo_ValoresSEIn_i}}}")
        #dfIn_str=dfIn_str.replace(f"{r"2.733"} ± {r"0.231"}", f"\\textbf{{{r"2.733"} ± {r"0.231"}}}")
    return(dfIn_str)
    
def func_quitarNAN(dfIn_str):
    dfIn_str=dfIn_str.replace("nan ± nan", 'NAN')
    return(dfIn_str)
```
