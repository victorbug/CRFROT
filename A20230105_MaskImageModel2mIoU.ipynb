```python
if __name__ == "__main__":

    import numpy as np

    # Ground truth segmentation mask
    ground_truth = np.array([[0, 0, 0],
                             [1, 0, 1]])

    # Predicted segmentation mask
    predicted = np.array([[0, 0, 0],
                          [0, 0, 1]])

    # Calculate the number of classes
    num_classes = np.max(ground_truth)

    # Initialize variables for class-wise intersection and union
    class_intersection = np.zeros(num_classes)
    class_union = np.zeros(num_classes)

    # Compute class-wise intersection and union
    lst_iou=[]
    for class_id in range(num_classes+1):
        print("class_id: ", class_id)
        gt_mask = (ground_truth == class_id)
        pred_mask = (predicted == class_id)

        class_intersection = np.sum(np.logical_and(gt_mask, pred_mask))
        class_union = np.sum(np.logical_or(gt_mask, pred_mask))

        # Calculate class-wise IoU
        class_iou = class_intersection / (class_union + 1e-10)  # Adding a small epsilon to avoid division by zero

        # Calculate class frequencies
        class_frequency= np.sum(ground_truth == class_id)/(ground_truth.size)
        #class_frequency = np.sum(ground_truth == np.arange(num_classes), axis=(0, 1))
        print("class_frequency: ", class_frequency, "class_iou: ", class_iou)

        # Calculate the frequency-weighted mIoU
        fw_miou = class_frequency*class_iou
        #fw_miou = np.sum(class_iou * class_frequency) / np.sum(class_frequency)

        print("Frequency-Weighted mIoU:", fw_miou)
        lst_iou.append(fw_miou)

    print("mIoU total: ", sum(lst_iou))
```
