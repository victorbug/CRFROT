```python
import numpy as np
from scipy.stats import wilcoxon

# Set a random seed for reproducibility
np.random.seed(42)

# Generate two sets of data points that are statistically significant
data_before = np.random.normal(loc=50, scale=10, size=100)  # Data before treatment
data_after = np.random.normal(loc=60, scale=10, size=100)  # Data after treatment

# Create a scatter plot to visualize the data
plt.scatter(np.arange(len(data_before)), data_before, label='Before Treatment', alpha=0.5)
plt.scatter(np.arange(len(data_after)), data_after, label='After Treatment', alpha=0.5)
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.legend()
plt.title('Data Before and After Treatment')
plt.show()


# Perform the Wilcoxon signed-rank test
statistic, p_value = wilcoxon(data_before, data_after)

# Output the results
print("Wilcoxon Test Statistic:", statistic)
print("P-Value:", p_value)

alpha = 0.05  # Set your significance level

if p_value < alpha:
    print("The difference is statistically significant")
else:
    print("The difference is not statistically significant")
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Create a sample numpy array
data = np.array([
    ['Value1', 'a', 'b', 99],
    ['Value1', 'a', 'x', 90],
    ['Value2', 'c', 'b', 85],
    ['Value3', 'f', 'd', 91]
])

# Extract column1 and column4
column1 = data[:, 0]
column4 = data[:, 3].astype(float)

# Encode column1 with numerical labels
label_encoder = LabelEncoder()
column1_encoded = label_encoder.fit_transform(column1)

# Calculate correlation coefficient
correlation = np.corrcoef(column1_encoded, column4)[0, 1]

# Perform linear regression
X = np.ones((len(column1_encoded), 2))
X[:, 1] = column1_encoded
coefficients = np.linalg.lstsq(X, column4, rcond=None)[0]
p_value = coefficients[1]

# Assess usefulness of column1
if abs(correlation) > 0.5 and p_value < 0.05:
    print("column1 is useful in the regression.")
else:
    print("column1 is not useful in the regression.")
```

```python
import pandas as pd

# Create a DataFrame with the given string representations of tuples
data = {'column_name': [
                        #"(1, 1), (1, 1), (1, 1)",
                        "(2, 1), (2, 1), (2, 1)",
                        "(3, 1), (3, 1), (3, 1)",
                        "(1, 1), (1, 1), (1, 1)",
                        "(5, 1), (5, 1), (5, 1)",
                        "(15, 1), (15, 1), (15, 1)"
                       ]}
df = pd.DataFrame(data)


def func_obtenerPrimeraTuplaIgual(dfIn):
    # Extract and clean the tuple values
    dfIn['column_name'] = dfIn['column_name'].str.extract(r'\((.*?)\)')
    dfIn['column_name'] = '(' + dfIn['column_name'] + ')'

    # Split the string into two columns and convert to integers
    dfIn[['column_1', 'column_2']] = dfIn['column_name'].str.strip('()').str.split(', ', expand=True)
    dfIn['column_1'] = dfIn['column_1'].astype(int)
    dfIn['column_2'] = dfIn['column_2'].astype(int)

    # Add a new column that checks if column_1 and column_2 are equal
    dfIn['equal_columns'] = dfIn['column_1'] == dfIn['column_2']
    if dfIn['equal_columns'].idxmax() != 0:
        print("The first occurrence of equal columns is not in the first row.")



    # Get the first elements of column_1 and column_2 where equal_columns is True
    first_equal_values = dfIn.loc[df['equal_columns'], ['column_1', 'column_2']].iloc[0]
    # Convert the first_equal_values to a string of the form "(x, y)"
    first_equal_values_str = f"({first_equal_values['column_1']}, {first_equal_values['column_2']})"
    
    first_equal_values_str = f'{first_equal_values_str}, {first_equal_values_str}, {first_equal_values_str}'
    # Print the first equal values
    #print(first_equal_values)
    return(first_equal_values_str)

func_obtenerPrimeraTuplaIgual(df)
```
