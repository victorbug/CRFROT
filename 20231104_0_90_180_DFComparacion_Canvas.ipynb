```python
def func_wilcoxon(arrIn1, arrIn2):
    import pandas as pd
    from scipy.stats import wilcoxon

    boo_out=None
    
    if all(arrIn2 - arrIn1 == 0):
        #print("All differences are zero. The Wilcoxon test cannot be performed.")
        boo_out="Diff equal to zero"
        return(boo_out)
    # Perform the Wilcoxon signed-rank test
    statistic, p_value = wilcoxon(arrIn1, arrIn2)

    # Output the results
    #print(f"Wilcoxon statistic: {statistic}")
    #print(f"P-value: {p_value}")

    # Interpret the results
    alpha = 0.05  # Set your desired significance level
    if p_value < alpha:
        #print("Reject the null hypothesis: There is a significant difference between 'before' and 'after' samples.")
        boo_out=True
    else:
        #print("Fail to reject the null hypothesis: There is no significant difference between 'before' and 'after' samples.")
        boo_out=False
    return(boo_out)
```
