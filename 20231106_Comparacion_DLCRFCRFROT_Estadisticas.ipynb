```python

def func_read_CSV_Pipeline5(str_dfIn):  

    def func_LessEqualMore(int_in1, int_in2):
        if ( int_in1 == int_in2 ):
            return("Mantiene")
        elif ( int_in1 > int_in2 ):
            return("Mejora")
        elif ( int_in1 < int_in2 ):
            return("Empeora")


    df_out = pd.read_csv(str_dfIn)#, header=None, skiprows=1)

    if("neuron"==
       "neuron"):#Neuronas se concentran en el medio asi que se usan otras discontinuidades. Aca les vamos a cambiar el nombre porque ntodo el escript esta configurado con discontinuidades 50, 100 150 y 200


        df_out.columns=df_out.columns.str.replace('175', '50')
        df_out.columns=df_out.columns.str.replace('225', '100')
        df_out.columns=df_out.columns.str.replace('275', '150')
        df_out.columns=df_out.columns.str.replace('325', '200')

    df_out=df_out.drop(columns=['lst_int_rotaciones'])#, 'str_imPathRaw', 'int_mostrarPrimeraIteracion', 'int_mostrarPrimeraIteracionRaw', 'str_nombrearchivoIn'])

    df_out=df_out.rename(columns={"nda_2DXX_ProbDLmap_ArgmaxRow50": "nda_2DXX_ProbDLmap_ArgmaxInRow50",
                                  "nda_2DXX_ProbDLmap_ArgmaxRow100": "nda_2DXX_ProbDLmap_ArgmaxInRow100",
                                  "nda_2DXX_ProbDLmap_ArgmaxRow150": "nda_2DXX_ProbDLmap_ArgmaxInRow150",
                                  "nda_2DXX_ProbDLmap_ArgmaxRow200": "nda_2DXX_ProbDLmap_ArgmaxInRow200",
                                 }
                        )#Esto es porque cambie una lesera y ahora son asi


    # Define the maximum length for column names


    df_out['str_PairwiseoKernel']=df_out['str_PairwiseoKernel'].replace('GaussianoSmoothnessKernel', 'Gaus')
    df_out['str_PairwiseoKernel']=df_out['str_PairwiseoKernel'].replace('BilateraloAppearanceKernel', 'Bil')
    df_out['str_PairwiseoKernel']=df_out['str_PairwiseoKernel'].replace('GaussianoSmoothnessKernelYBilateraloAppearanceKernel', 'GausYBil')


    #df_out['dict_tpl_ParametroXYIn22']=df_out['dict_tpl_ParametroXYIn22'].replace('Cuadrado', '')
    df_out['dict_tpl_ParametroXYIn22'] = df_out['dict_tpl_ParametroXYIn22'].str.replace("""{'Cuadrado':""", '')
    df_out['dict_tpl_ParametroXYIn22'] = df_out['dict_tpl_ParametroXYIn22'].str.replace("""'RectanguloVertical90':""", '')
    df_out['dict_tpl_ParametroXYIn22'] = df_out['dict_tpl_ParametroXYIn22'].str.replace("""'RectanguloHorizontal0':""", '')
    df_out['dict_tpl_ParametroXYIn22'] = df_out['dict_tpl_ParametroXYIn22'].str.replace("""}""", '')


    
    if("paraMetricasDice"==False):
        """
        Una manera muy poco elegante de computar DICE, haciendo pasar esas columnas como miou para lo cual ya esta todo listo. DESACTIVAR DESPUES DE UTILIZARLO!
        """
        
        if("wholeImg"==
          "wholeImg"):
            if("eliminarOriginales"==
              "eliminarOriginales"):
                df_out=df_out.rename(columns={"flo64_ProbDLmap_Argmax_miou2": "PARADICELADEJOFUERATEMPORALMENTE1",
                                              "flo64_ProbCRFmap_Argmax_miou": "PARADICELADEJOFUERATEMPORALMENTE2",
                                              "flo64_ProbCRFROTmap_Argmax_miou": "PARADICELADEJOFUERATEMPORALMENTE3"
                                             }
                                    )

            if("reemplazarConColumnasDice"==
              "reemplazarConColumnasDice"):
                df_out=df_out.rename(columns={"flo64_ProbDLmap_Argmax_Dice2": "flo64_ProbDLmap_Argmax_miou2",
                                              "flo64_ProbCRFmap_Argmax_Dice": "flo64_ProbCRFmap_Argmax_miou",
                                              "flo64_ProbCRFROTmap_Argmax_Dice": "flo64_ProbCRFROTmap_Argmax_miou"
                                             }
                                     )

        #DISCONTINUIDADES
        if("discontinuidades"==
          "discontinuidades"):
            
            if("eliminarOriginales"==
              "eliminarOriginales"):
                df_out=df_out.rename(columns={"nda_2DXX_ProbDLmap_ArgmaxInRow50": "PARADICELADEJOFUERATEMPORALMENTE4",
                                          "nda_2DXX_ProbDLmap_ArgmaxInRow100": "PARADICELADEJOFUERATEMPORALMENTE5",
                                          "nda_2DXX_ProbDLmap_ArgmaxInRow150": "PARADICELADEJOFUERATEMPORALMENTE6",
                                          "nda_2DXX_ProbDLmap_ArgmaxInRow200": "PARADICELADEJOFUERATEMPORALMENTE7",
                                         }
                                     )
                df_out=df_out.rename(columns={"nda_2DXX_ProbCRFmap_ArgmaxRow50": "PARADICELADEJOFUERATEMPORALMENTE8",
                                          "nda_2DXX_ProbCRFmap_ArgmaxRow100": "PARADICELADEJOFUERATEMPORALMENTE9",
                                          "nda_2DXX_ProbCRFmap_ArgmaxRow150": "PARADICELADEJOFUERATEMPORALMENTE10",
                                          "nda_2DXX_ProbCRFmap_ArgmaxRow200": "PARADICELADEJOFUERATEMPORALMENTE11",
                                         }
                                     )
                df_out=df_out.rename(columns={"nda_2DXX_ProbCRFROTmap_ArgmaxRow50": "PARADICELADEJOFUERATEMPORALMENTE12",
                                          "nda_2DXX_ProbCRFROTmap_ArgmaxRow100": "PARADICELADEJOFUERATEMPORALMENTE13",
                                          "nda_2DXX_ProbCRFROTmap_ArgmaxRow150": "PARADICELADEJOFUERATEMPORALMENTE14",
                                          "nda_2DXX_ProbCRFROTmap_ArgmaxRow200": "PARADICELADEJOFUERATEMPORALMENTE15",
                                         }
                                     )
                

            if("reemplazarConColumnasDice"==
              "reemplazarConColumnasDice"):
                #Mas arriba ya se reemplazo en neuronas las posiciones (175,225, 275 y 325)
                df_out=df_out.rename(columns={"nda_2DXX_ProbDLmap_ArgmaxDiRow50": "nda_2DXX_ProbDLmap_ArgmaxInRow50",
                                              "nda_2DXX_ProbDLmap_ArgmaxDiRow100": "nda_2DXX_ProbDLmap_ArgmaxInRow100",
                                              "nda_2DXX_ProbDLmap_ArgmaxDiRow150": "nda_2DXX_ProbDLmap_ArgmaxInRow150",
                                              "nda_2DXX_ProbDLmap_ArgmaxDiRow200": "nda_2DXX_ProbDLmap_ArgmaxInRow200",
                                             }
                                     )

                df_out=df_out.rename(columns={"nda_2DXX_ProbCRFmap_ArgmaxDiRow50": "nda_2DXX_ProbCRFmap_ArgmaxRow50",
                                              "nda_2DXX_ProbCRFmap_ArgmaxDiRow100": "nda_2DXX_ProbCRFmap_ArgmaxRow100",
                                              "nda_2DXX_ProbCRFmap_ArgmaxDiRow150": "nda_2DXX_ProbCRFmap_ArgmaxRow150",
                                              "nda_2DXX_ProbCRFmap_ArgmaxDiRow200": "nda_2DXX_ProbCRFmap_ArgmaxRow200",
                                             }
                                     )

                df_out=df_out.rename(columns={"nda_2DXX_ProbCRFROTmap_ArgmaxDiRow50": "nda_2DXX_ProbCRFROTmap_ArgmaxRow50",
                                              "nda_2DXX_ProbCRFROTmap_ArgmaxDiRow100": "nda_2DXX_ProbCRFROTmap_ArgmaxRow100",
                                              "nda_2DXX_ProbCRFROTmap_ArgmaxDiRow150": "nda_2DXX_ProbCRFROTmap_ArgmaxRow150",
                                              "nda_2DXX_ProbCRFROTmap_ArgmaxDiRow200": "nda_2DXX_ProbCRFROTmap_ArgmaxRow200",
                                             }
                                     )
                
                
        if("dejarloEnPorcentajePorquepuedeInfluir"==
          "dejarloEnPorcentajePorquepuedeInfluir"):
            #df_out["flo64_ProbDLmap_Argmax_miou2"]=df_out["flo64_ProbDLmap_Argmax_miou2"]*100
            #df_out["flo64_ProbCRFmap_Argmax_miou"]=df_out["flo64_ProbCRFmap_Argmax_miou"]*100
            #df_out["flo64_ProbCRFROTmap_Argmax_miou"]=df_out["flo64_ProbCRFROTmap_Argmax_miou"]*100
        
            columns_to_multiply = [
                                        'flo64_ProbDLmap_Argmax_miou2',
                                        'flo64_ProbCRFmap_Argmax_miou',
                                        'flo64_ProbCRFROTmap_Argmax_miou',
                
                                        "nda_2DXX_ProbDLmap_ArgmaxInRow50",
                                        "nda_2DXX_ProbDLmap_ArgmaxInRow100",
                                        "nda_2DXX_ProbDLmap_ArgmaxInRow150",
                                        "nda_2DXX_ProbDLmap_ArgmaxInRow200",
                
                                        "nda_2DXX_ProbCRFmap_ArgmaxRow50",
                                        "nda_2DXX_ProbCRFmap_ArgmaxRow100",
                                        "nda_2DXX_ProbCRFmap_ArgmaxRow150",
                                        "nda_2DXX_ProbCRFmap_ArgmaxRow200",
                
                                        "nda_2DXX_ProbCRFROTmap_ArgmaxRow50",
                                        "nda_2DXX_ProbCRFROTmap_ArgmaxRow100",
                                        "nda_2DXX_ProbCRFROTmap_ArgmaxRow150",
                                        "nda_2DXX_ProbCRFROTmap_ArgmaxRow200",
                                    ]

            df_out[columns_to_multiply] *= 100

    #display_full(df_out[["flo64_ProbDLmap_Argmax_miou2", 
    #                     "flo64_ProbCRFmap_Argmax_miou", 
    #                     "flo64_ProbCRFROTmap_Argmax_miou",
    #                     "nda_2DXX_ProbDLmap_ArgmaxInRow50",
    #                     "nda_2DXX_ProbDLmap_ArgmaxInRow100",
    #                     "nda_2DXX_ProbDLmap_ArgmaxInRow150",
    #                     "nda_2DXX_ProbDLmap_ArgmaxInRow200"
    #                    ]].head(2))
    
    #print("muchambre")
    
    return(df_out)




```

```python
def func_wilcotomaronce(group, str_methodBase, str_methodPropuesto):
#def calculate_mean_difference(group, str1, str2):
    #return (group[str1] - group[str2]).mean()
    from scipy.stats import wilcoxon

    # Example paired data for Method A and Method B
    #method_a = [14, 15, 16, 18, 19]
    #method_b = [12, 14, 16, 17, 20]

    # Performing the one-sided Wilcoxon signed-rank test (alternative='greater')
    statistic, p_value = wilcoxon(group[str_methodPropuesto], group[str_methodBase], alternative='greater')

    # Display the test statistic and p-value
    print(f"One-sided Wilcoxon signed-rank statistic: {statistic}")
    print(f"P-value: {p_value}")

    # Interpret the results based on the p-value and significance level
    alpha = 0.05  # Significance level
    if p_value < alpha:#El pvalor es cuanto le creemos a la hipotesis nula, por lo tanto generalmente queremos que sea bajo
        print("Reject the null hypothesis: str_methodPropuesto tends to produce larger values than str_methodBase.")
    else:
        print("Fail to reject the null hypothesis: No sufficient evidence that str_methodPropuesto produces larger values than str_methodBase.")
    return(p_value)
```

```python
import pandas as pd
import io

# Sample LaTeX string representing a DataFrame
latex_string = r'''\begin{tabular}{lll}
\toprule
{} & Column1 & Column2 \\
\midrule
0 &       A &       X \\
1 &       A &       Y \\
2 &       B &       X \\
3 &       B &       Y \\
4 &       B &       Z \\
\bottomrule
\end{tabular}'''

# Convert LaTeX string to DataFrame
# First, extract the table content from the LaTeX string
table_content = latex_string.split('\n')[3:-2]  # Extract rows from the LaTeX string

# Join the extracted content back into a string and read it using Pandas read_csv
df = pd.read_csv(io.StringIO('\n'.join(table_content)), delimiter='\t')

print(df)
```
